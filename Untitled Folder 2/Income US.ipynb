# <a id='1'>1. Introduction</a>

This data set was extracted from the  US Census Bureau Database. The Census Bureau provides data about people and economy of US. The census records data specific to individual respondents are not available to the public until 72 years after a given census was taken, but aggregate statistical data derived from the census are released as soon as they are available. Every census up to and including 1940 is currently available to the public and can be viewed on microfilm released by the National Archives and Records Administration, the official keeper of archived federal census records.

The current national census has been in progress since 2020; the next census is scheduled for 2030. Since 2013, the Census Bureau began discussions on using technology to aid data collection starting with the 2020 census.For years between the decennial censuses, the Census Bureau issues estimates made using surveys and statistical models, in particular, the Population Estimates Program and American Community Survey.

# <a id='4'>4. Data manipulation and validation</a>

#### Preparing all libraries

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
sns.set()
```

##### Importing Data

```python
df = pd.read_csv('adult.csv')
                   
```

##### Changing column names

```python
df.columns = (['age','workclass','fnlwgt','education','educationNum',
                                  'maritalStatus','occupation','relationship','race','sex','capitalGain',
                                  'capitalLoss','hoursPerWeek','nativeCountry','income'])
```

#### Misisng value in the dataset are denote with the string “?”. So I replace it with Nan to make the treatment of missing values easier.

```python
df.replace("?", np.nan, inplace = True)
```

#### Income is showns as "<=50K" and ">50K" so i replaced it with 0 and 1 to make calculations easier

```python
df.replace("<=50K", 0, inplace = True)
df.replace(">50K", 1, inplace = True)
```

#### Dataset statistics

```python
df.describe()
```

#### Changing some data values

```python
df.replace("Federal-gov", "FederalGov", inplace = False)
df.replace("Local-gov", "LocalGov", inplace = True)
df.replace("Never-worked", "NeverWorked", inplace = True)
df.replace("Self-emp-inc", "selfEmployedIncome", inplace = True)
df.replace("Self-emp-not-inc", "selfEmployedNotIncome", inplace = True)
df.replace("State-gov", "StateGovernment", inplace = True)
df.replace("Without-pay", "WithoutPay", inplace = True)
```

```python
df.replace("United-States", "UnitedStates", inplace = True)
df.replace("El-Salvador", "ElSalvador", inplace = True)
df.replace("Puerto-Rico", "PuertoRico", inplace = True)
df.replace("Dominican-Republic", "DominicanRepublic", inplace = True)
df.replace("Trinadad&Tobago", "TrinadadTobago", inplace = True)
df.replace("Outlying-US(Guam-USVI-etc)", "OutlyingUS", inplace = True)
df.replace("Holand-Netherlands", "HolandNetherlands", inplace = True)
```

### Graph shows how education affects workclass

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig_dims = (15, 8)
fig, ax = plt.subplots(figsize=fig_dims)
hist = sns.histplot(ax=ax,data = df, x = "education", hue = "workclass", multiple = "stack"); #plotting histogram 
plt.xticks(rotation = 90); #rotating X-axis labels by 90 degrees for readability 
plt.title("Workclass education distribution", size = 20); #title of the graph
```

We can see that there is more people who have only school education, but those work for themselves

### Level of income of different nationalities

```python
fig = plt.figure(figsize=(15, 4))
df.groupby('nativeCountry')['income'].mean().sort_values().plot(kind='bar', color='pink')
plt.xticks(rotation=90)
plt.xlabel("Native Country")
plt.ylabel('Income')
plt.show()
```

We see that there are lot of immigrants. And we can suggest that the survey was taken in little cities
