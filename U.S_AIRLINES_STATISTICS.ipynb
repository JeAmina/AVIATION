```python

df = pd.read_excel('DATASET.xlsx')
df['Transportation Employment - Air Transportation'] = df['Transportation Employment - Air Transportation'].str.replace(',', '')
df['Transportation Employment - Air Transportation'] = df['Transportation Employment - Air Transportation'].astype(int)
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'].str.replace(',', '')
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'].str.split('.', expand=True)
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - Domestic'].astype(int)
#df['DOMESTIC FLIGHTS'] = df['DOMESTIC FLIGHTS'].str.replace(',', '')
df['DOMESTIC FLIGHTS'] = df['DOMESTIC FLIGHTS'].astype(int)
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'].str.replace(',', '')
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'].str.split('.', expand=True)
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'].astype('int64')
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'].astype('int64')
df
```

How employee number has changed after COVID-19 quarantine

```python
df1=(df['Transportation Employment - Air Transportation'].shift() - df['Transportation Employment - Air Transportation'])/df['Transportation Employment - Air Transportation'].shift() * 100
df1.plot()
```

How flights has changed after COVID-19 quarantine

```python
df1=(df['DOMESTIC FLIGHTS'].shift() - df['DOMESTIC FLIGHTS'])/df['DOMESTIC FLIGHTS'].shift() * 100
df1.plot()
```

Retrieving loading factor as Passengers per Flights

```python
relationfatalitie= df["Air Safety - Air Carrier Fatalities"] / df["Air Safety - General Aviation Fatalities"]
relationfatalities = pd.DataFrame(relationfatalitie)
df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'] = df['U.S. Air Carrier Cargo (millions of revenue ton-miles) - International'].astype('int64')

df = df.assign(relationfatalities=relationfatalities.values)
df2 = df[['Air Safety - Air Carrier Fatalities', 'Air Safety - General Aviation Fatalities', 'relationfatalities']]
df2
```

Retrieving percentage of air carrier per air fatalities

```python
loadFactor= df["PASSENGERS"] / df["DOMESTIC FLIGHTS"]
loadFactors = pd.DataFrame(loadFactors)
df['PASSENGERS'] = df['PASSENGERS'].astype(float)


df = df.assign(loadFactors=loadFactors.values)
df2 = df[['PASSENGERS', 'DOMESTIC FLIGHTS', 'loadFactors']]
df2
```
